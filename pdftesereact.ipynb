```python
import pandas as pd 
import numpy as np
from PIL import Image
import io
from pdf2image import convert_from_path
import pytesseract
```

```python
def extract_text_from_pdf(pdf_path):
    pages = convert_from_path(pdf_path=pdf_path)
    text_data = ''
    for page in pages:
        text = pytesseract.image_to_string(page)
        text_data += text + '/n'
    return text_data

text = extract_text_from_pdf('/home/sanjay/Downloads/Trinidad and tobago/805_India.pdf')
```
